```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import the turnstile_180901.txt csv
df = pd.read_csv('turnstile_180901.txt')

print(df.shape)
df.head()
```

```python
# Rename all the columns to lower case
df.columns = [column.lower() for column in df.columns]

df.head()
```

```python
# More 'pythonic' way to write loop
df.columns = [column.lower() for column in df.columns]
# wrap entire right side in [] to indicate list
df.head()
```

```python
# Change the index to 'linename'

df = df.set_index('linename')
# must remember to assign 'df' to the operation

df.head()
```

```python
# reset the index
df.reset_index().head() # preview looks good
```

```python
df = df.reset_index()
df.head()
```

```python
# Create another column 'Num_Lines' that is a count of 
# how many lines pass through a station. 
# Then sort your DataFrame by this column in descending order.

# Create new 'Num_Lines' column
df['Num_Lines'] = 0
df.head()
```

```python
# count of characters in 'linename'
# .map() to 'Num_Lines'

# remember 'text' is the argument in the lambda function
# len(text) is the operation to map to each entry
df['Num_Lines'] = df['linename'].map(lambda text: len(text))
df.head()
```

```python
# write a function to clean column names
# make lower case and no underscores or spaces

def clean(col_name):
    cleaned = col_name.lower().replace(' ', '').replace('_','')
    return cleaned
```

```python
# test cleaning function
example = 'RANDOM_COL NAME'
clean(example)
```

```python
type(df.columns)
```

```python
# method using a for loop
df.columns = [clean(column) for column in df.columns]
```

```python
df.columns
```

```python
# method using .map() function 
# probably a lot better this way
df.columns = df.columns.map(clean)

# df.columns = [df.columns.map(clean)] creates a multiIndex array (dont do this)
```

```python
df.columns
```

```python
# Convert data type of 'date' column to a date
# df['numlines']
# pd.to_datetime(df['date']).head() # this generates an error b/c df.columns is a multi index
pd.to_datetime(df['date']).head() # this works 
```

```python
# make change official
df.date = pd.to_datetime(df.date)
```

```python
df.date.head()
```

```python
# add new column 'day_of_week' that represents day of the week

# use datetime functions
df['date'].dt.day_name().head() #preview looks good
```

```python
df['day_of_week'] = df.date.dt.day_name()
```

```python
# group the data by day of week and plot the sum of the numeric columns
df.groupby('day_of_week').sum()
```

```python
# DataFrame object
type(df.groupby('day_of_week').sum())
```

```python
# plot the sum of the numeric columns
x = df.groupby('day_of_week').sum()

# .plot can be run on a series or dataframe object
x.plot(kind='barh') # 'barh' means horizontal bar plot

plt.show()
```

```python
# Remove the index of grouped or 'x'
x = x.reset_index() #preview looks good

# Print the first five rows of grouped
x.head()
```

```python
# Add a new column 'is_weekend' that maps the 'day_of_week' column 
# using the dictionary weekend_map

weekend_map = {'Monday': 0, 'Tuesday': 0, 'Wednesday': 0, 'Thursday': 0, 'Friday': 0, 'Saturday': 1, 'Sunday': 1}

df['is_weekend'] = df['day_of_week'].map(weekend_map) # mapping dictionary values
```

```python
# preview dataframe
df.loc[0:,['day_of_week','is_weekend']]
```

```python
# Group the data by weekend/weekday and plot the sum of the numeric columns
df.groupby('is_weekend').sum() #preview looks good
```

```python
y = df.groupby('is_weekend').sum()
y.plot(kind = 'barh')
```
